```python
from IPython.display import clear_output

class RentalAddresses:
    def __init__(self):
        self.rentalAddress = rentalAddress 
        self.rentalType = rentalType

    def rentalInfo(self):
        self.rentalAddress = input("What is the address of the property you would like to rent out? ")
        self.rentalType = input("What type of home is this property? (apartment/condo/house/duplex/etc) ")

class RentalNumbers:
    def __init__(self):
        #income from house 
        self.rentalInc = 0
        self.laundry = 0
        self.storage = 0
        self.miscIncome = 0 
        #expenses 
        self.proptaxes = 0
        self.insurance = 0
            # utilities 
        self.utilities = 0
        self.electric = 0
        self.water = 0
        self.sewer = 0 
        self.gas = 0
        self.garbage = 0 
        self.HOA = 0
        self.LawnSnow = 0 
        self.vacancy = 0 
        self.repairs = 0 
        self.CapitalExpenses = 0
        self.propertymanagement = 0 
        self.mortgage = 0
        self.otherExp = 0
        # investment
        self.propertyCost = 0 
        self.downPayment = 0
        self.closingCosts = 0
        self.rehab = 0
        self.miscInvestment = 0
        #totals 
        self.totalInvestment = 0
        self.totalExpenses = 0 
        self.totalIncome = 0
    
    def Income(self): 
            print("        -------------------Income-------------------")
            try: 
                self.rentalInc = int(input("How much will you be charging monthly for Rent? "))
                self.laundry = int(input("How much will you be charging for Laundry? (washer/dryer) "))
                self.storage = int(input("How much will you be charging for Storage for the tenant(s)? "))
                self.totalIncome = (self.rentalInc+self.laundry+self.storage)
                print(f"\n=========================================================\nYour total income on this property is calculated at ${self.totalIncome}\n=========================================================")

            except ValueError:
                print(f"Please answer as Numbers") 
                
            
    def Expenses(self):
        print("        -------------------Expenses-------------------")
        try: 
            self.proptaxes = int(input("How much will Property Taxes be?" ))
            self.insurance = int(input("How much will Insurance be?" ))
            # utilities 
            print("             ----------Utilities----------")
            self.electric = int(input("Electricity: "))
            self.water = int(input("Water: "))
            self.sewer = int(input("Sewage: "))
            self.gas = int(input("Gas: "))
            self.garbage = int(input("Garbage: "))
            self.LawnSnow =int(input("Lawn/Snow Maintenance: ")) 
            self.utilities = (self.electric+self.water+self.sewer+self.gas+self.garbage+self.LawnSnow)
            print("             -------End of Utilities-------")
            self.HOA =int(input("How much are monthly HOA charges? ")) 
            self.vacancy =int(input("How much will Vacancy cost? ")) 
            self.repairs =int(input("How much will repairs cost? "))  
            self.CapitalExpenses =int(input("How much will you be paying in Capital Expenses? ")) 
            self.propertymanagement =int(input("How much will you pay for Property Management? "))  
            self.mortgage =int(input("How much is the MONTHLY Mortgage payment? ")) 
            self.otherExp = (self.HOA+self.vacancy+self.repairs+self.CapitalExpenses+self.propertymanagement+self.mortgage)
            
            self.totalExpenses = (self.proptaxes+self.insurance+self.utilities+self.otherExp)
            print(f"\n=========================================================\nYour total expenses on this property is calculated at ${self.totalExpenses}\n=========================================================")
        except ValueError:
            print(f"Please answer as Numbers") 
            
    def Investments(self):
        print("        -------------------Investments-------------------")
        try: 
            self.propertyCost = int(input("How much does the property cost? ")) 
            self.downPayment = int(input("How much are you putting down as a Down Payment? "))
            self.closingCosts = int(input("How much are the Closing Costs? "))
            self.rehab = int(input("How much will repairing/rehabilitating the property cost? "))
            self.miscInvestment = int(input("Please enter an estimate for all other misc costs for investment on this property. "))
            
            self.totalInvestment = (self.propertyCost+self.downPayment+self.closingCosts+self.rehab+self.miscInvestment)
            print(f"\n=========================================================\nYour total investments on this property is calculated at ${self.totalInvestment}\n=========================================================")
        except ValueError:
            print(f"Please answer as Numbers")    

       
    
class RentalCalc:
    def CashFlowCalc(self):
        self.cashFlow = (self.totalIncome - self.totalExpenses)*12
        print(f"\n=========================================================\nYour Annual Cashflow is ${self.cashFlow}.")
    

    def ROICalc(self):
        self.ROI = (self.cashFlow / self.totalInvestment)*100.00
        print(f"\nYour Calculated Return on investment at {self.rentalAddress.title()} is {self.ROI:.2f}%\n=========================================================")

    
    
    

    
class Main:
    @classmethod
    def run(self):
        print("\n====================\nLet's git ROI wit it\n====================")
        while True:
            RentalAddresses.rentalInfo(self)
            RentalNumbers.Income(self)
            RentalNumbers.Expenses(self)
            RentalNumbers.Investments(self)
            RentalCalc.CashFlowCalc(self)
            RentalCalc.ROICalc(self)
            editCalc = input("Would you like to edit any of your figures? yes/no ")
            while editCalc.lower() == "yes":
                editSection = input("Which section would you like to edit? (investments/income/expenses/address/exit) ")
                if editSection.lower() == "investments":
                    RentalNumbers.Investments(self)
                    RentalCalc.CashFlowCalc(self)
                    RentalCalc.ROICalc(self)
                elif editSection.lower() == "income":
                    RentalNumbers.Income(self)
                    RentalCalc.CashFlowCalc(self)
                    RentalCalc.ROICalc(self)
                elif editSection.lower() == "expenses":
                    RentalNumbers.Expenses(self)
                    RentalCalc.CashFlowCalc(self)
                    RentalCalc.ROICalc(self)
                    break
                elif editSection.lower() == "exit":
                    break
            startOver = input("Would you like to calculate the ROI on another property? yes/no ")
            if startOver.lower() == "yes":
                clear_output()
                continue
            elif startOver.lower() == "no":
                print(f"Have a great day! Or not, your choice.")
                break 
```

```python
Main.run()
```
